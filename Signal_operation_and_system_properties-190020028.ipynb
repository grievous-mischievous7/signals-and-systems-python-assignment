```python
#QN 1
import numpy as np
import matplotlib.pyplot as plt

n=np.arange(-5,6,1)

def plot(x):
    x_timeinverse=x[::-1]
    plt.stem(n,x,use_line_collection=True)
    plt.title('x[n]');plt.show()
    
    x_odd=1/2*(x-x_timeinverse)
    plt.stem(n,x_odd,use_line_collection=True)
    plt.title('x_o[n]');plt.show()
    
    x_even=1/2*(x+x_timeinverse)
    plt.stem(n,x_even,use_line_collection=True)
    plt.title('x_e[n]');plt.show()
    
    x=x_even+x_odd
    plt.stem(n,x,use_line_collection=True)
    plt.title('x[n]');plt.show()
    

#1)a
x=np.zeros((11))
for i in range(11):
    if n[i]>=0:
        x[i]=1

plot(x)

```

```python
# 1)b
n=np.arange(-5,6,1)
x=np.zeros((11))
for i in range(11):
    if n[i]>=0:
        x[i]=(1/2)**(i-5)
        
print(x)
plot(x)
```

```python
#1)c
x=np.cos(.25*(np.pi)*n)
plot(x)
```

```python
#QN 2)A
x=np.zeros((11))
for i in range(11):
    if n[i]>=0:
        x[i]=1
n0=2
y=np.zeros(11)
for i in range(11):
    y[i]=np.sum(x[i-n0:i+n0+1])
print(x)
print(y)

plt.stem(n,x,use_line_collection=True)
plt.title('x[n]');plt.show()
plt.stem(n,y,use_line_collection=True)
plt.title('y[n]');plt.show()
```

```python
# QN 3)A
n=np.arange(-500,501,1)

x=np.zeros(1001)
x[500+4]=3
plt.stem(n[500:520],x[500:520],use_line_collection=True)
plt.title('x[n]');plt.show()

u=np.zeros(1001)
u[501:505]=1

h=np.zeros(1001)
for i in range(1,5):
    h[500+i]=.75**(i)*u[500+i]
plt.stem(n[500:520],h[500:520],use_line_collection=True)
plt.title('h[n]');plt.show()

y=np.zeros(1001)
for i in range(1001):
    for k in range(1001):
        if x[k]*h[i-k]!=0 :
            y[i-500]+=x[k]*h[i-k]
plt.stem(n[500:520],y[500:520],use_line_collection=True)
plt.title('y[n]');plt.show()
```
