```python
import numpy as np
import matplotlib.pyplot as plt
n=np.arange(-7,8)
#QN1 part1 
x_n=np.zeros(15)
for i in n:
    if i>=0:
        x_n[i+7]+=1/pow(2,i)
    else:
        x_n[i+7]+=1/pow(2,-i)
x_z=np.zeros(15)
for i in n:
    if i>=1 or i<=-1:
        x_z[i+7]+=(2*i)/(2*i-1)
    elif i==0:
        x_z[0+7]=0  
plt.stem(n,x_n,use_line_collection = True)
plt.title('x[n]');plt.xlabel('n')
plt.xticks(range(-7,8,1))
plt.show()

plt.stem(n,x_z,use_line_collection = True)
plt.title('X(Z)');plt.xlabel('z')
plt.xticks(range(-7,8,1))
plt.show()   
```

```python
#QN1 part 2
n=np.arange(0,7)
x=np.zeros(7)
for i in n:
    x[i+0]=1/pow(4,i)
y=np.zeros(7)
for i in n:
    if i==0:
        y[i+0]= ((-2/3)*pow(-0.5,0)) + ((1/6)*pow(0.25,0))
    else:
        y[i+0] = (x[i]-y[i-1])/2
plt.stem(n,x,use_line_collection = True)
plt.title('x[n]');plt.xlabel('n')
plt.xticks(range(0,7,1))
plt.show()

plt.stem(n,y,use_line_collection = True)
plt.title('y[n]');plt.xlabel('n')
plt.xticks(range(0,7,1))
plt.show()        
```

```python
#QN2 
n=np.arange(0,20)
x=np.zeros(20)
for i in n:
    x[i]= 1 + np.sin((np.pi*3*i/8)+np.pi/4)
y=np.zeros(20)
for i in n:
    y[i]= np.sin((np.pi*3*i/8)+np.pi/4)

plt.stem(n,x,use_line_collection = True)
plt.title('x[n]');plt.xlabel('n')
plt.show()

plt.stem(n,y,use_line_collection = True)
plt.title('y[n]');plt.xlabel('n')
plt.show() 

    

```

```python
#QN3 part 1
n = np.linspace(-50,50,500)
x=[]
for i in n.tolist():
     x.append(np.sin(150*np.pi*i))
plt.title("x(t)")
plt.xlabel("t")
plt.plot(n,np.asarray(x))
plt.show()       
```
