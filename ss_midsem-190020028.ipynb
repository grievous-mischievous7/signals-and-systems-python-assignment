```python
#6)B
from math import e
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=(15,5)
n=np.linspace(-15,15)
y=np.cos(np.pi*n/10)
plt.xlabel('Values of n')
plt.title('$cos(n*\pi/10)$')
ax = plt.axes()
ax.set_xlim([-15,15])
ax.set_ylim([-1.5, 1.5])
ax.stem(n,y,use_line_collection=True)
```

```python
n=np.arange(-2,8,1)
x=np.zeros(10)
ax = plt.axes()
ax.set_xlim([-1,5])
ax.set_ylim([-1.5,1.5])
plt.xlabel('Values of n')
plt.title('$e^-{2n}u[n]$')
for i in range(10):
    if n[i]>=0:
        x[i]=pow(e,-2*(i-2))
plt.stem(n,x,use_line_collection=True)
```

```python
n=np.arange(-20,20)
x=np.zeros(40)
ax =plt.axes()
ax.set_xlim([-15,15])
ax.set_ylim([-70,70])
plt.xlabel('n')
plt.title('y[n]')
for i in range(40):
    x[i]=pow(e,4) * np.cos(np.pi*i/10)-pow(e,2) * np.cos(np.pi/10*(i+1))/(pow(e,4)-2*pow(e,2)*np.cos(np.pi/10)+1)
plt.stem(n,x,use_line_collection=True)
```

```python
#QN 4)
n=np.arange(-2,2,1/20)
f= 1 + 2*np.cos(2*np.pi *n) + np.sin(2*np.pi + np.pi/2)
plt.stem(n,f,use_line_collection=True)

y_real=[1.5 , 0 , 1,  0, 1.5]
y_imag=[ -.5, .5 , 0, -0.5 , 0.5]
x=[-2,-1,0,1,2]
```

```python
ax = plt.axes()
ax.grid()
ax.set_xlim([-3,3])
ax.set_ylim([0,2])
plt.xlabel("k");
plt.ylabel('Real($a_{k}$)');
plt.title("Real Part of Fourier Coefficient")
ax.stem(x,y_real,use_line_collection=True)
```

```python
ax = plt.axes()
ax.grid()
ax.set_xlim([-3,3])
ax.set_ylim([-1,1])
plt.xlabel("'k', the integer harmonic");
plt.ylabel('Imag(Ak)');
plt.title("Imaginary Part of Fourier coefficient (*j)")
ax.stem(x,y_imag,use_line_collection=True)
```
