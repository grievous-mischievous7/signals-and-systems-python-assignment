```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def grad(x1, ALPHA , x, y):    # Initial x, Learning Rate, X axis, Y axis
        x2 = x1 
        dy = np.diff(y)
        dy = np.append(dy ,0)
        j=0            
        xnew= []
        ynew= []
        i = 0
        for i in range(0,len(x)):
                if abs(x[i] - x2) <= 0.05:
                    index = i
       
        while abs(dy[index]) > 0.01 and j<1000 :   
            for i in range(0,len(x)):
                if abs(x[i] - x2) <= 0.05:
                    index = i
            
            x2 = x2 + ALPHA*(dy[index])
            j+=1
            xnew.append(x[index])
            ynew.append(y[index])

    
        
        plt.plot(x,y)
        plt.plot(xnew, ynew)
        plt.show()
```

```python
mean = 0
sigma=1
x= np.arange(-5, 5, 0.2)
y=[]
y=1/(sigma*math.sqrt(2*math.pi)) * np.exp(-1/2 * ((x-mean)/sigma)**2)
plt.title('Gaussian function');
grad(-1, 0.5, x, y)
```

```python
dy = np.diff(y)
dy = np.append(dy ,0)
plt.plot(x,dy)
plt.title('deltaG:Gaussian function')
```
