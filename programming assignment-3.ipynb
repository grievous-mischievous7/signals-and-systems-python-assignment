# QN 1

```python
import numpy as np
import matplotlib.pyplot as plt
import math
#Part 1
n=np.arange(-10,10)
x=np.zeros(21)
x[20] = 1 
plt.stem(n,x,use_line_collection = True)
plt.title('x[n]');plt.xlabel('n')
plt.show()
y=np.zeros(41)
for i in n:
    if i>=0:
        y[i+20] = 1/(i+1)*(x[i+20] + i*y[i-1 +20])
    
plt.stem(n,y,use_line_collection = True)
plt.title('y[n]');plt.xlabel('n')
plt.show()
#Part 2 
n=np.arange(-20,21)
x=np.zeros(41)
x[20+1] = 1

plt.stem(n,x,use_line_collection = True)
plt.title('x[n]');plt.xlabel('n')
plt.show()
y=np.zeros(41)
for i in n:
    if i>=0:
        y[i+20] += 0.25*y[i-1+20] + x[i+20]

plt.stem(n,y,use_line_collection = True)
plt.title('y[n]');plt.xlabel('n')
plt.show()
```

```python
n=np.arange(-20,21)
x=np.zeros(41)
for i in n:
    if(i%4==0):
        x[i]=4
    elif(i%4==1):
        x[i]=8
plt.title('X[n]')
plt.xlabel('n')
plt.stem(n,x,use_line_collection=True)
```

```python
k=np.arange(-20,21)
real_ak=np.zeros(41)
for i in k:
    real_ak[i]=1+2*np.cos(i*np.pi/2)
imag_ak=np.zeros(41)
for j in k:
    imag_ak[j]=-2*np.sin(j*np.pi/2)
mod_ak=np.zeros(41)
for e in k:
    mod_ak[e]=math.sqrt(5+4*np.cos(e*np.pi/2))
plt.title('real a_k')
plt.xlabel('k')
plt.stem(k,real_ak,use_line_collection=True)
plt.show()

plt.title('imag a_k')
plt.xlabel('k')
plt.stem(k,imag_ak,use_line_collection=True)
plt.show()

plt.title('real a_k')
plt.xlabel('k')
plt.stem(k,mod_ak,use_line_collection=True)
plt.show()
```
