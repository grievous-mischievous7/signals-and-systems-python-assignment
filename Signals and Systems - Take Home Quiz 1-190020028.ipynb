```python
#QN 1)a
import numpy as np

import matplotlib.pyplot as plt

fs=40

for fa in [5,15,20,25,35,40,45,60]:
    
    fd=fa/fs
    
    n=np.arange(40)
    
    x=np.cos(2*(np.pi)*fd*n)
    
    plt.title(f'X[n]=cos({2*fd} $\pi$ n)')
    
    plt.xlabel('n')
    
    plt.ylabel('X[n]')
    
    plt.stem(n,x,use_line_collection=True)
    
    plt.show()
    




```

QN 1)b 

we can see from the plots that with increase in analog frequency $f_a$ the frequency of oscillation is not necessarily   increasing,(Though frequency of oscillation increases initially with analog frequency upto a particular value of $f_a$,because the frequency of oscillation of $f_{a}=5$ and that of $f_{a}=45$ is same,and after $f_{a}=40$ the frequency of oscillations repeats of that of initial frequency of oscillation,i.e,after $f_{a}=f_{s}$ frequency of oscillation repeats previous values.
This is due to periodic property of this DT function.

example: $cos(1.0 \pi n) =  cos(3.0 \pi n)$  for $f_{a}=20$ and $f_{a}=60$ respectively.

```python
#QN 2)a
import numpy as np

import matplotlib.pyplot as plt

ts=0.05

for n0 in [2,3,5,10]:
    
    n=np.arange(30)

    x=np.sin(2*(np.pi)*ts*(n-n0))
    
    plt.title(f'X[n]=cos( {2*ts} $\pi$ (n-{n0}) )')
    
    plt.xlabel('n')
    
    plt.ylabel('X[n]')
    
    plt.stem(n,x,use_line_collection=True)
    
    plt.show()
```
