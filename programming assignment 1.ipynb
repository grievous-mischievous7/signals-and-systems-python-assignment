```python
# qn. a) 1
import matplotlib.pyplot as plt
import numpy as np

print ("enter  Ts and f \n")
Ts=int(input())
f =int(input())

n=np.linspace(-5,6,100)
x=np.cos(2*(np.pi)*f*n*Ts)
plt.title('X[n]=cos(2$\pi$fnTs)')
plt.xlabel('n')
plt.ylabel('X[n]')
plt.stem(n,x,use_line_collection=True)
```

```python
# qn. a) 2
n=np.linspace(-1,1,60)
x=1*(abs(n)<0.5)
plt.title('Rectangle')
plt.xlabel('n')
plt.ylabel('X[n]')
plt.stem(n,x,use_line_collection=True)
```

```python
# qn a) 3
from scipy import signal 
print("enter information of % of rise edge")
rise=float(input())
n=np.linspace(0,0.5,40)
plt.title('Triangle')
plt.xlabel('n')
plt.ylabel('X[n]')
x=signal.sawtooth(rise*np.pi*2*n,0.5)
plt.stem(n,x,use_line_collection=True)
```

```python
# qn a) 4

print ("enter Ts and f \n")
Ts=int(input())
f =int(input())
n=np.linspace(-5,5,100)
x=np.exp(-n*Ts)*np.cos(2*n*np.pi*f*Ts)
plt.title('X[n]=exp^(-nTs)*cos(2$\pi$fnTs)')
plt.xlabel('n')
plt.ylabel('X[n]')

plt.stem(n,x,use_line_collection=True)
```
